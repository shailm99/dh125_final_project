```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
institutions = pd.read_csv('Most-Recent-Cohorts-Institution.csv')
institutions.columns
```

```python
pd.options.display.max_columns = 999
institutions
```

```python
colleges = institutions[['OPEID6', 'INSTNM', 'CITY', 'STABBR', 'ZIP','MENONLY', 'WOMENONLY', 'ADM_RATE_ALL', 'SAT_AVG_ALL', 'HIGHDEG',
'HBCU', 'PBI','HSI', 'NANTI', 'CCUGPROF', 'CCSIZSET', 
'UGDS_WHITE', 'UGDS_BLACK', 'UGDS_HISP', 'UGDS_ASIAN', 'COSTT4_A', 'AVGFACSAL', 'C150_4', 
'C150_4_WHITE', 'C150_4_BLACK', 'C150_4_HISP', 'C150_4_ASIAN', 'FEMALE', 
'MN_EARN_WNE_P10', 'MN_EARN_WNE_MALE0_P10', 'MN_EARN_WNE_MALE1_P10', 'GT_THRESHOLD_P10']]
colleges.to_csv('colleges.csv')
```

```python
field = pd.read_csv('Most-Recent-Cohorts-Field-of-Study.csv')
field
```

```python
unique_bachelors = field[field.CREDLEV == 3].groupby(['OPEID6', 'CONTROL'])['CIPDESC'].count().reset_index()
```

```python
colleges = colleges.join(unique_bachelors, on='OPEID6', lsuffix='L', rsuffix='R').drop(columns = ['OPEID6R'])
```

```python
colleges['Co-Ed'] = (colleges['MENONLY'] == 0) & (colleges['WOMENONLY'] == 0)
```

```python
colleges.rename(columns = {'CIPDESC' : 'Number of Unique Majors Offered (Bachelors'}, inplace = True)
```

```python
colleges.to_csv('colleges.csv')
```
